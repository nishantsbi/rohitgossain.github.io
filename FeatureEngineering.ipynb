```python
from featuretools.primitives import make_agg_primitive, make_trans_primitive
from featuretools.variable_types import Text, Numeric
```

```python
def word_count(column):
    word_counts = []
    for value in column:
        words = value.split(None)
        words_counts.append(len(words))
    return words_counts
```

```python
WordCount = make_trans_primitive(function=word_count,input_types=[Text], return_type=Numeric)
```

```python
feature_defs = ft.dfs(entityset=es,target_entity="sessions", agg_primitives=[Mean],trans_primitives=[WordCount])
```

```python
feature_defs
```
