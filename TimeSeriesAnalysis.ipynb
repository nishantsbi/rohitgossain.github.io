```r
w = rnorm(500,0,1)
plot.ts(w)
```

```r
v = filter(w, sides = 2, rep(1/3,3))
plot.ts(v)
```

```r
w = rnorm(550,0,1)
x = filter(w,filter = c(1,-9),method = 'recursive')
plot.ts(x)
```
